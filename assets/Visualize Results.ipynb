```python
import numpy as np 
import matplotlib.pyplot as plt
import os
```

```python
LIST = ['0028','0374','0017']
EXT = '.png'
for name in LIST:
    plt.figure(figsize=(15,5))
    plt.subplot(1,4,1)
    plt.imshow(plt.imread(os.path.join('org', name + EXT)))
    plt.axis('off')
    plt.subplot(1,4,2)
    plt.imshow(plt.imread(os.path.join('wb', name + '_out' + EXT)))
    plt.axis('off')
    plt.subplot(1,4,3)
    plt.imshow(plt.imread(os.path.join('irb', name + '_out' + EXT)))
    plt.axis('off')
    plt.subplot(1,4,4)
    plt.imshow(plt.imread(os.path.join('irgf', name + '_out' + EXT)))
    plt.axis('off')
    plt.show()
```

```python
EXT = '.jpg'
name = '23'
plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.imshow(plt.imread(os.path.join('org', name + EXT))[100:300,1000:])
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(plt.imread(os.path.join('wb', name + '_out' + EXT))[100:300,1000:])
plt.axis('off')
plt.subplot(1,4,3)
plt.imshow(plt.imread(os.path.join('irb', name + '_out' + EXT))[100:300,1000:])
plt.axis('off')
plt.subplot(1,4,4)
plt.imshow(plt.imread(os.path.join('irgf', name + '_out' + EXT))[100:300,1000:])
plt.axis('off')
plt.show()
```
